```
import numpy as np
import matplotlib.pyplot as plt
```

```
#1

speech = open('speechs.txt', 'r')
text = speech.read()
text = text.lower()
words = text.split()

word_cnt = {}

processed_words = []

for word in words:
  alphabets = list([c for c in word if c.isalpha()])
  word = ''.join(alphabets)
  if (word != ''):
    if (word in word_cnt):
      word_cnt[word] += 1
    else:
      word_cnt[word] = 1
    processed_words.append(word)

for word in list(word_cnt.keys()):
  print(word, ': ', word_cnt[word])
```

```
#2

idx = {}
rev_idx = {}

i = 0

for word in processed_words:
  if (word not in idx):
    idx[word] = i
    rev_idx[i] = word
    i += 1

#2148 - Number of unique words (after processing)

matrix = np.zeros((2148, 2148))
for num in range(0, len(processed_words) - 1):
  i = idx[processed_words[num]]
  j = idx[processed_words[num + 1]]
  matrix[i][j] += 1

def freq_next(s):
  if (s not in idx):
    print('Error: No such word exists in the text file.')
  else:
    print(matrix[idx[s]][:])

print('Enter a word:')
s = input()
freq_next(s)
```

```
#3

def generate_sentence(s, cnt):
  sentence = []
  if (s in idx):
    sentence.append(s)
    cnt -= 1
    w = s
    while (cnt):
      i = idx[w]
      limit = np.random.uniform(0, 1) * word_cnt[w]
      found = 0
      index = 0
      cfreq = 0
      while (index < 2148):
        cfreq += matrix[i][index]
        if (cfreq >= limit):
          found = 1
          break
        else:
          index += 1
      if (found):
        sentence.append(rev_idx[index])
        w = rev_idx[index]
      else:
        break
      cnt -= 1
    print(*sentence)
  else:
    print('Error: The word you entered does not exist in the file.')

print('Enter a string (to start your text): ')
s = input()
print('Enter the length of your message (in words): ')
length = int(input())
generate_sentence(s, length)
```
