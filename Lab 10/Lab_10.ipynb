```python
A = [1, 2, 3, 4, 6]

p_out_min = [0.01, 0.02, 0.03, 0.10, 0.00]
p_out_max = [0.10, 0.20, 0.30, 0.50, 0.00]

def p_out(w, action_idx):
  return p_out_max[action_idx] + (p_out_min[action_idx] - p_out_max[action_idx]) * (w - 1) / 9

best_score = [[0 for j in range(11)] for i in range(301)]
best_shot = [[-1 for j in range(11)] for i in range(301)]

#computing the values for all states
for i in range(1, 301):
  for j in range(1, 11):
    for k in range(5):
      score = (1 - p_out(j, k)) * (A[k] + best_score[i - 1][j]) + p_out(j, k) * best_score[i - 1][j - 1]
      if (score > best_score[i][j]):
        best_score[i][j] = round(score, 2)
        best_shot[i][j] = A[k]
```

```python
for i in range(301):
  print(best_score[i])

for i in range(301):
  print(best_shot[i])
```
