```
#finding the squared distance between two points
def distance(pt1, pt2):
  return np.linalg.norm(np.subtract(pt1, pt2))
print(distance(np.array([1, 1]), np.array([0, 0])))
```

```
#creating a random set of points using 2(a)
n = int(input('Enter the size of the dataset you want to generate: '))
dataset = create_data_set(n)
print('Randomly generated dataset:\n', dataset)

#2(b) printing out the k closest points
def closest(dataset, k, pt):
  sz = dataset.shape[0]
  a = np.empty([sz, 2])
  for i in range(sz):
    a[i][0] = distance(pt, dataset[i, :])
    a[i][1] = i
  sorted_a = a[np.argsort(a[:, 0])]
  if (k > sz): 
    print('Error: the dataset is not large enough!')
  result = np.empty([min(k, sz), 2])
  for i in range(min(k, sz)):
    result[i] = dataset[int(sorted_a[i][1]), :]
  return(result)

#accepting input
print('Enter the spaced coordinates of the new point:')
x_new, y_new = map(float, input().split())
new_pt = np.array([x_new, y_new])
print('How many of the closest points do you want?')
k = int(input())

#printing the result
print(closest(dataset, k, new_pt))
```

```
#creating a random set of points using 2(a)
n = int(input('Enter the size of the dataset you want to generate: '))
dataset = create_data_set(n)
print('Randomly generated dataset:\n', dataset)

#2(c) printing out k points in the' dataset that make a positive angle with the point pt
def positive_angle(dataset, k, pt):
  result = []
  i = 0
  while (len(result) < k and i < dataset.shape[0]):
    if (np.dot(dataset[i, :], pt) >= 0):
      result.append(dataset[i, :])
    i += 1
  if (len(result) < k):
    print('Error: not enough points exist!')
  result = np.array(result)
  return result

print('Enter the spaced coordinates of the new point:')
x_new, y_new = map(float, input().split())
new_pt = np.array([x_new, y_new])
print('How many of the closest points do you want?')
k = int(input())
print(positive_angle(dataset, k, new_pt))
```

```
#3(a)
n = int(input('Enter the size of the dataset you want to generate: '))
dataset = create_data_set(n)
print('Enter the spaced coordinates of the new point:')
x_new, y_new = map(float, input().split())
new_pt = np.array([x_new, y_new])
print('How many of the closest points do you want?')
k = int(input())
b = closest(dataset, k, new_pt)
plt.scatter(dataset[:, 0], dataset[:, 1], color = 'b')
plt.scatter(b[:, 0], b[:, 1], color = 'r')
```

```
#4
x = np.random.uniform(-0.5, 0.5, size = 100)
prefix_sums = np.empty([100])
k = np.empty([100]) 
for i in range(100):
  prefix_sums[i] = x[i]
  if (i):
    prefix_sums[i] += prefix_sums[i - 1]
  k[i] = i + 1
plt.plot(k, prefix_sums / k, 'r', label = 'y')
plt.plot(k, (1 / k) ** (1 / 2), 'g', label = 'f1')
plt.plot(k, (-1) * ((1 / k) ** (1 / 2)), 'b', label = 'f2')
plt.title('Plot for Q4')
plt.xlabel('k')
plt.ylabel('y - axis')
plt.legend()
```
